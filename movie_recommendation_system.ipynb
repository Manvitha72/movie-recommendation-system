```python
movies = movies[['id', 'title', 'overview', 'genre']]
```

```python
movies['tags'] = movies['overview'] + movies['genre']
```

```python
new_movies = movies.drop(columns=['overview', 'genre'])
```

```python
new_movies
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000, stop_words='english')
```

CountVectorizer is a class from the sklearn.feature_extraction.text module in the scikit-learn library. This class is used to convert text data into a matrix of token counts. It transforms a corpus of text documents into a matrix where each row represents a document and each column represents a token (word).

The max_features parameter specifies the maximum number of features (i.e., words) to consider.

The stop_words parameter specifies a list of stop words to remove from the text data before converting it to a matrix.

```python
cv
```

```python
vector = cv.fit_transform(new_movies['tags'].values.astype('U')).toarray()
```

new_movies['tags'].values converts the pandas Series (the tags column) to a NumPy array.
.astype('U') ensures that the data type of the array is Unicode strings.

fit_transform() is a method that:
  Fits the CountVectorizer to the text data, learning the vocabulary of the 10,000 most frequent tokens (after removing stop words).
  Transforms the text data into a sparse matrix where each row represents a document and each column represents a token from the learned vocabulary.

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

cosine_similarity(vector) calculates the cosine similarity between all pairs of documents represented in the vector matrix.

It is defined as the cosine of the angle between two vectors. The formula is:
cosine similarity= A⋅B / ∥A∥∥B∥

```python
similarity
```

```python
new_movies[new_movies['title']=="The Godfather"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_movies.iloc[i[0]].title)
```

```python
def recommend(movies):
    index = new_movies[new_movies['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_movies.iloc[i[0]].title)
```

```python
recommend("Batman")
```
